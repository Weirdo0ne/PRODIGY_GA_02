```python
import torch
from diffusers import StableDiffusionPipeline
import matplotlib.pyplot as plt
```

```python
model_id = "CompVis/stable-diffusion-v1-4"
pipe = StableDiffusionPipeline.from_pretrained(model_id)
pipe = pipe.to("cuda")  # Ensure the model runs on the Gpu
```

```python
prompt = input("To Generate img, tpye here :- ")

# Generate the images
num_images = 4
images = pipe([prompt] * num_images).images  # Adjust to extract images from the output

# Function to plot images
def plot_images(images):
    plt.figure(figsize=(20, 20))
    for i in range(len(images)):
        ax = plt.subplot(1, len(images), i + 1)
        plt.imshow(images[i])
        plt.axis("off")
    plt.show()

# Plot the images
plot_images(images)
```
